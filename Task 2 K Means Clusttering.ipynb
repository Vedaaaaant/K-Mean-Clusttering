# Importing Libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Importing Data

```python
from sklearn import datasets
df = datasets.load_iris()
iris = pd.DataFrame(df.data, columns = df.feature_names)
iris.head() 
```

# EDA

```python
sns.pairplot(iris,palette='coolwarm')
```

```python
plt.figure(figsize=(12,8))
sns.jointplot(x='sepal length (cm)',y='sepal width (cm)',data=iris,kind='scatter',color='green')
```

```python
plt.figure(figsize=(12,8))
sns.jointplot(x='petal length (cm)',y='petal width (cm)',data=iris,kind='scatter',color='red')
```

```python
plt.figure(figsize=(12,8))
sns.distplot(iris['petal length (cm)'],bins=5)
```
